## Funciones

```python
def extract_TRMM_area(extent, time_ini, time_fin, inPath, outPath, formato='NetCDF',
                     plot=True):
    """Genera un archivo 'NetCDF' o 'csv' con los datos de precipitación TRMM para una zona y periodo de estudio.
    
    Parámetros:
    -----------
    extent:      list [W, E, S, N]. Extensión de la zona de interés
    time_ini:    string. Fecha inicial en formato '%Y-%m-%d-%H'
    time_fin:    string. Fecha final en formato '%Y-%m-%d-%H'
    inPath:        string. Ruta donde se encuentran los archivos de entrada
    outPath: string. Ruta donde guardar los archivos de salida
    formato:     string. Tipo de archivo de salida a exportar: 'NetCDF' o 'csv'
    plot:        boolean. Si mostrar un ejemplo de mapa de precipitación
    
    Salidas:
    --------
    Si el formato es 'NetCDF', el archivo 'Precipitacion_TRMM.nc' en la carpeta 'outPath'.
    Si el formato es 'csv', los archivos 'Precipitacion_TRMM.nc' y 'Coordendas_TRMM.csv' en la carpeta 'outPath'.
    """
    
    # definir pasos temporales
    time = pd.date_range(start=time_ini, end=time_fin, freq='3h')
    
    # EXTRAER COORDENADAS
    # -------------------
    # cargar el día inicial de la serie, por ejemplo
    datetimeobject = datetime.strptime(time_ini,'%Y-%m-%d-%H')
    newformat = datetimeobject.strftime('%Y%m%d.%H')
    # leer el archivo NetCDF
    if time[0].year > 1999 and time <= datetime.strptime('2010-09-30', '%Y-%m-%d'):
        root_grp = Dataset(inPath + '3B42.' + newformat + '.7A.HDF.nc4')
    else:
        root_grp = Dataset(inPath + '3B42.' + newformat + '.7.HDF.nc4')
    # extraer variables
    lon = root_grp.variables['nlon'][:]
    lat = root_grp.variables['nlat'][:]
    precp = root_grp.variables['precipitation'][:].data.T
    precp[precp <= 0] = np.nan # corregir valores negativos de precipitación
    data_masked = np.ma.array(precp, mask=np.isnan(precp))
    # generar coordenadas
    pos_x = np.where((lon >= extent[0]) & (lon <= extent[1]))[0]
    pos_y = np.where((lat >= extent[2]) & (lat <= extent[3]))[0]
    [XX, YY] = np.meshgrid(lon, lat)
    [XX_s, YY_s] = np.meshgrid(np.sort(lon[pos_x]), np.sort(lat[pos_y]))
    Coordendas = pd.DataFrame(index=np.arange(0,len(pos_x) * len(pos_y)),
                              columns=['Lon', 'Lat'])
    Coordendas.iloc[:,0] = np.reshape(XX_s.T, (-1,1))
    Coordendas.iloc[:,1] = np.reshape(YY_s.T, (-1,1))
    
    # mapa con la precipitación en el día inicial de la serie
    if plot:
        fig, ax = plt.subplots(nrwos=1, ncols=1, figsize=(20,15),
                               subplot_kw=dict(projection=ccrs.PlateCarree()))
        cs = ax.pcolormesh(XX, YY, data_masked.data, vmin=np.min(dato[0]),
                           vmax=np.max(dato[0]), cmap=plt.get_cmap("cool"))
        bf = 10 # buffer
        ax.set_extent([extent[0] - bf, extent[1] + bf, extent[2] - bf, extent[3] + bf])
        ax.coastlines(resolution='50m')
        ax.add_feature(cartopy.feature.BORDERS, linestyle='-')
        ax.plot(XX_s.reshape(-1,1), YY_s.reshape(-1,1), '.r', markersize=5)
        cbar = fig.colorbar(cs, ax=ax, orientation='horizontal', pad=0.015)
        cbar.set_label('Precipitacion mm/h')
 
    # EXTRAER DATOS
    # ---------------
    # objetos donde guardar los datos
    if formato == 'csv':
        Precipitacion = []
    elif formato == 'NetCDF':
        Precipitacion_TRMM = np.zeros((len(time), len(pos_y), len(pos_x)))
    
    # carga de datos
    for t, i in enumerate(tqdm.tqdm(time)):
        datetimeobject = datetime.strptime(str(i), '%Y-%m-%d %H:%M:%S')
        newformat = datetimeobject.strftime('%Y%m%d.%H')
        # leer archivo NetCDF
        if i.year > 1999 and i.year <= 2010:
            if i.year == 2010 and i.month > 9:
                ncfile = '3B42.{0}.7.HDF.nc4'.format(newformat)
                if os.path.isfile(inPath + ncfile):
                    root_grp = Dataset(inPath + ncfile)
                else:
                    print(ncfile)
                    continue
            else:
                ncfile = '3B42.{0}.7A.HDF.nc4'.format(newformat)
                if os.path.isfile(inPath + ncfile):
                    root_grp = Dataset(inPath + ncfile)
                else:
                    print(ncfile)
                    continue
        else:
            ncfile = '3B42.{0}.7.HDF.nc4'.format(newformat)
            if os.path.isfile(inPath + ncfile):
                root_grp=Dataset(inPath + ncfile)
            else:
                print(ncfile)
                continue
                
        # guardar datos de precipitación
        if formato == 'csv':
            Precp = np.reshape(root_grp.variables['precipitation'][pos_x, pos_y].data,
                               (1,-1))
            Precp[Precp < 0] = np.nan
            Precipitacion.append(np.array(Precp, dtype='float')[0])
        elif formato == 'NetCDF':
            Precipitacion_TRMM[t,:,:] = root_grp.variables['precipitation'][pos_x,pos_y].data.T

    # EXPORTAR DATOS
    # --------------
    if formato == 'csv':
        # exportar los datos de precipitación a partir de un 'data frame'
        Precipitacion_TRMM = pd.DataFrame(Precipitacion, index=pd.date_range(start=time_ini, end=time_fin, freq='3h'),
                                          columns=np.arange(0, len(pos_x) * len(pos_y)))
        Precipitacion_TRMM.to_csv(outPath + 'Precipitacion_TRMM.csv')
        # exportar coordenadas
        Coordendas.to_csv(outPath + 'Coordendas_TRMM.csv')
    elif formato == 'NetCDF':
        # corregir datos erróneos
        Precipitacion_TRMM[Precipitacion_TRMM < 0] = -9999
        # Creamos netCDF
        nc = Dataset(outPath + 'Precipitacion_TRMM.nc', 'w', format='NETCDF4')
        # Global Attributes
        nc.description= 'TRMM precipitation data'
        # nc dimensions
        nc.createDimension('lon', len(pos_x))
        nc.createDimension('lat', len(pos_y))
        nc.createDimension('time', len(time))
        # crear variables
        xx_nc=nc.createVariable('lon', 'float64', ('lon'))
        yy_nc=nc.createVariable('lat', 'float64', ('lat'))
        time_nc=nc.createVariable('time', 'float64', ('time'))
        variable_1_nc=nc.createVariable('Precipitacion', 'float64',
                                        ('time', 'lat', 'lon'))
        #units
        xx_nc.units = 'degrees_east'
        yy_nc.units = 'degrees_north'
        time_nc.units = 'hours since ' + time[0].strftime('%Y-%m-%d %H:%M:%S')
        variable_1_nc.units = 'mm'
        #long_name
        xx_nc.long_name = 'longitude coordinate'
        yy_nc.long_name = 'latitude coordinate'
        time_nc.long_name = 'day of the year'
        variable_1_nc.long_name = '3-hourly precipitation'
        #fill_value
        variable_1_nc.missing_value = '-9999'
         # rellenar variables
        time_nc[:] = np.arange(0, 3*len(time), 3)
        xx_nc[:] = lon[pos_x]
        yy_nc[:] = lat[pos_y]
        variable_1_nc[:] = Precipitacion_TRMM[:]
        nc.close()
```
