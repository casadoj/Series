```python
def download_ERA5_copernicus(date,extension,variable,path_output):
    from datetime import datetime, timedelta
    import cdsapi
    d=datetime.strptime(date,'%Y-%m-%d %H:%M:%S')
    if variable=='total_precipitation':
        file_name='Precipitacion_'+str(d.year)+'_'+"{0:0=2d}".format(d.month)+'_'+"{0:0=2d}".format(d.day)+'.nc'
    elif variable=='2m_temperature':
        file_name='Temperatura_'+str(d.year)+'_'+"{0:0=2d}".format(d.month)+'_'+"{0:0=2d}".format(d.day)+'.nc'
    if os.path.exists(path_output+file_name)==True:
        return
    else:
        c = cdsapi.Client()
        c.retrieve(
            'reanalysis-era5-land',
            {
                'format': 'netcdf',
                "area": extension,
                'variable': variable,
                'month': "{0:0=2d}".format(d.month),
                'day': "{0:0=2d}".format(d.day),
                'year': str(d.year),
                'time': [
                    '00:00', '01:00', '02:00',
                    '03:00', '04:00', '05:00',
                    '06:00', '07:00', '08:00',
                    '09:00', '10:00', '11:00',
                    '12:00', '13:00', '14:00',
                    '15:00', '16:00', '17:00',
                    '18:00', '19:00', '20:00',
                    '21:00', '22:00', '23:00',
                ],
            },
            path_output+file_name)
```

```python
import pandas as pd
import tqdm
import os
for i in tqdm.tqdm(pd.date_range(start='1981-01-01',end='2020-02-26',freq='D')):
    download_ERA5_copernicus(str(i),'-19.00/-63.00/-29.00/-53.00','total_precipitation','/mnt/e/BID_ASUNCION/Clima/ERA5/')
    download_ERA5_copernicus(str(i),'-19.00/-63.00/-29.00/-53.00','2m_temperature','/mnt/e/BID_ASUNCION/Clima/ERA5/')
```
